### i have created a simple data locally where it contains img of actor i have 4 class (4 different actors images)

```python
import os
import torch
from torch.utils.data import DataLoader, random_split
from torchvision import datasets , transforms
import matplotlib.pyplot as plt
```

```python
dataset_folder = './ImgFolderDataset'


# Define transformations (resizing, converting to tensor, normalization)
transform = transforms.Compose([
    transforms.Resize((80, 80)),  # Resize images to 32x32 (or use any other size)
    transforms.ToTensor(),
    transforms.Normalize(mean=[0.5, 0.5, 0.5], std=[0.5, 0.5, 0.5])  # Normalize to [-1, 1]
])
```

```python
# Load the dataset using ImageFolder
dataset = datasets.ImageFolder(root=dataset_folder, transform=transform)
```

```python
train_data = DataLoader(dataset,batch_size=10,shuffle=True)
```

```python
data = iter(train_data)
imgage, label = next(data)
```

```python
# Display 8 images
import matplotlib.pyplot as plt

fig, axes = plt.subplots(2, 4, figsize=(10, 5))
for i, ax in enumerate(axes.flat):
    img = imgage[i].permute(1, 2, 0).numpy()  # Convert to HWC for matplotlib
    img = (img * 0.5) + 0.5  # De-normalize
    ax.imshow(img)
    ax.set_title(f"Label: {dataset.classes[label[i]]}")  # Use class name instead of label
    ax.axis('off')
plt.tight_layout()
plt.show()
```
