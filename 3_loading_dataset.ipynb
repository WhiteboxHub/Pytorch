```python
import torch
from torch.utils.data import DataLoader, Dataset
from torchvision import datasets,transforms
```

```python
transform = transforms.Compose([
    transforms.ToTensor(),
    transforms.Normalize((0.5,),(0.5,)),
    transforms.Resize((22,22))
])
```

# loading the MNIST dataset 

MNIST dataset is a pool if images which contain the numbers writen on it

```python
minst_train = datasets.MNIST(root='./data',train = True,transform=transform,download=True)
minst_test= datasets.MNIST(root='./data',train = False,transform=transform,download=True)
```

### lets see the sample of the MNIST dataset

```python
import matplotlib.pyplot as plt

def Sample_data_view(data):
    fig,ax = plt.subplots(3,3,figsize=(8,8))
    index = 0
    for i in range(0,3):
        for j in range(0,3):
            image, label = data[index]
            image = image.squeeze().numpy()
            ax[i,j].imshow(image,cmap='gray')
            index += 1
    plt.axis('off')
    plt.show()

Sample_data_view(minst_train)
```
